```python
## Split Height column as it is string (Making it foot and inches)

Height_split = df["Height"].str.split("-", n = 1, expand = True)
df["Feet"] = Height_split[0]
df["Inches"] = Height_split [1]
df.drop(columns = ["Height"], inplace = True)
df
```

```python
## creating new column with height in cm and making it int.

schema = {"Feet": np.uint8,
          "Inches": np.uint8}
df = df.astype(schema)
df["Height(cm)"] = df.apply(lambda x: x["Feet"] * 30.48 + x["Inches"] * 2.54, axis = 1)
df.drop(columns = ["Feet", "Inches"], inplace = True)
df
```

```python
df.loc[110:112, ["Height(cm)"]]
```

```python
## What is the weight in kg for 112th record (111th index)?

df["Weight(KG)"] = df.apply(lambda x: x["Weight"]*0.453592, axis = 1)
df.loc [111, ["Weight(KG)"]]
```

```python
##What is the BMI for 112th record (111th index)?

df["BMI"] = df.apply(lambda x : x["Weight(KG)"] / (x["Height(cm)"]/100)**2, axis = 1)
df.loc [111, "BMI"]
```

```python
df.drop(columns = ["Weight"], inplace = True)
df
```

```python
df["BMI"].max(), df["BMI"].min()
```

```python
## Create a new column with the name “BMI Evaluation”. Assign the BMI status based on the following condition,

## BMI <= 18.4 - "Underweight"
## BMI <= 24.9 - "Healthy"
## BMI <= 29.9 - "Over Weight"
## BMI <= 34.9 - "Severely Over Weight"
## BMI <= 39.9 - "Obese"
## 39.9 - "Severely Obese." 

def BMI_status (bmi) :
    if bmi <= 18.4 :
        return "Underweight"
    elif (bmi > 18.4 and bmi <= 24.9) :
        return "Healthy"
    elif (bmi > 24.9 and bmi <= 29.9) :
        return "Over Weight"
    elif (bmi > 29.9 and bmi <= 34.9) :
        return "Severly Over Weight"
    elif (bmi > 34.9 and bmi <= 39.9) :
        return "obese"
    else:
        return "Severely Obese"
df["BMI Evaluation"] = df.apply(lambda x: BMI_status(x["BMI"]), axis = 1)
df
```

```python
## Compute the total amount paid by each team to the players.
## What is the amount paid by Chicago Bulls to the players?

df.groupby(by = "Team")["Salary"].sum()
```

```python
## Compute the average age of each team by their playing position. 
## What is the average age of the Utah Jazz team players playing at the ‘PG’ position?

dx = df.groupby (by = ["Team", "Position"]).agg({"Age" : "mean"}).reset_index()
dx[(dx["Team"].isin(["Utah Jazz"])) & (dx["Position"].isin(["PG"]))]
```

```python
## Compute each team's average, maximum, and minimum age by their playing position. 
## What is the average, maximum, and minimum age of the Atlanta Hawks players playing at the ‘PG’ position?

dy = df.groupby (by = ["Team", "Position"])["Age"]. agg (["mean","max","min"]).reset_index()
dy[(dy["Team"].isin(["Atlanta Hawks"])) & (dx["Position"].isin(["PG"]))]
```

```python
## Compute each team's average age and total salary by their playing position. Sort the grouped results based on salary.

dz = df.groupby (by = ["Team", "Position"]).agg({"Age" : "mean", "Salary" : "sum"}).reset_index().sort_values(by = "Salary", ascending = False)
dz
```
