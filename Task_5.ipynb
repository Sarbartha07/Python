Here we have two lists (year and GDP). create a line plot with proper labels and title to the figure. Also add marker, color, size and line width.

Year = [1950, 1960, 1970, 1980, 1990]
GDP = [50.8, 65.9, 75.8, 92.6, 125.8]

```python
import matplotlib.pyplot as plt
import seaborn as sns

Year = [1950, 1960, 1970, 1980, 1990]
GDP = [50.8, 65.9, 75.8, 92.6, 125.8]

plt.figure(figsize = (10,2))
plt.plot(Year,GDP, marker = '*', color = 'red', linewidth = 2)
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("Year vs GDP")
plt.show()
```

Read mtcars dataset and create a histogram of mpg (mile per gallon) feature with proper lables and title. Give it an appropriate edge and face color. Add number of bins (15).

It has been claimed that manual cars have high mileage per gallon. Use mtcars dataset and create box plot between am and mpg columns and compare cylinders also in the same plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv('/content/mtcars - mtcars (1).csv')

df['transmission'] = df['am'].replace({0: "automatic", 1: "manual"})
df

box_plot = px.box(df, x = 'transmission', y = 'mpg', color = 'cyl', title='Mileage per Gallon (mpg) by Transmission Type and Number of Cylinders', labels={'mpg': 'Miles per Gallon (mpg)', 'transmission': 'Transmission Type'}, category_orders={'transmission': ['Automatic', 'Manual']})
box_plot.show()
```

Use mtcars dataset and create histogram of mpg column and scatterplot between mpg and weight using subplot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/mtcars - mtcars (1).csv")

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,5))
sns.histplot(df['mpg'], ax = ax[0])
sns.scatterplot(x = df['mpg'], y = df['wt'], ax = ax[1])
ax[0].set_title("Histogram")
ax[1].set_title("Scatter Plot")
plt.show()
```
