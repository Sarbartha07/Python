```python
## Create a genre-level report with metrics average rating, the average number of votes, and average revenue. 

df.groupby(by = "Genre").agg({"Rating": "mean", "Votes": "mean", "Revenue_millions": "mean"})
```

```python
## Create a report to showcase the revenue of each movie, 
## as % revenue concerning the total revenue of the respective genre and year of the movie.


##creating dataframe of Total_revenue Genre and Year wise
Total_Revenue = df.groupby (by = ["Genre", "Year"])["Revenue_millions"].sum().reset_index()
Total_Revenue.rename(columns = {"Revenue_millions" : "Total_Revenue"}, inplace = True)

## creating cleaned datafram of original dataframe
dfc = df.dropna(how = "any")

## Merge Total_Revenue dataframe with actual dataframe
Merge_df = pd.merge(dfc, Total_Revenue)

## creating new column of percent revenue with respective Genre and Year
Merge_df["Percent_Revenue"] = Merge_df.apply(lambda x : (x["Revenue_millions"] / x["Total_Revenue"]) * 100, axis = 1)
Merge_df[Merge_df["Title"] == "Split"]
```

```python
## Add a column â€˜Votes_normâ€™ in the IMDB dataset using apply() function where Votes_norm is 
##[Votes - min(Votes)]*10/[max(votes) - min(votes)


# Define a function to calculate 'Votes_norm'
def calculate_votes_norm(x):
    min_votes = df['Votes'].min()
    max_votes = df['Votes'].max()
    return (x['Votes'] - min_votes) * 10 / (max_votes - min_votes)

# Apply the function to create the 'Votes_norm' column
df['Votes_norm'] = df.apply(calculate_votes_norm, axis=1).round(2)
df["Votes_norm"].mean()
```

```python
## Add a column â€˜Total_ratingâ€™  in the IMDB dataset using apply() function where Total_rating is â€˜Ratingâ€™+ â€˜Votes_normâ€™.

df["Total_Rating"] = df.apply(lambda x : x["Rating"] + x["Votes_norm"], axis = 1)
df["Total_Rating"].max()
```

```python
## Create a new column â€˜Revenue_binsâ€™ so that â€˜Revenue_millionsâ€™ are categorized in buckets 
## 0-50, 51-100, 101-150, and 150+. Which bucket has the highest number of movies?

def revenue(rev) :
    if (rev >=0 and rev<= 50):
        return "0-50"
    elif (rev >= 51 and rev <= 100) :
        return "51-100"
    elif (rev >= 101 and rev <= 150) :
        return "101-150"
    else :
        return "150+"
df["Revenue_bins"] = df.apply(lambda x : revenue(x["Revenue_millions"]), axis = 1)
df["Revenue_bins"].value_counts()
```
